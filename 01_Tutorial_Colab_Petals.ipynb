```
canvas = Canvas_cv()      # create a canvas on which we can draw

canvas.draw_circle01(1)   # draw a circle

canvas.show()             # show the drawing on the canvas
```

```
n = 8  # divide the circle like a pizza, in n slices

def point_on_circle(canvas, p_of_n, n, draw = False):
  slice_angle = 360 / n
  angle = p_of_n * slice_angle
  point = canvas.get_circle_point01(angle)
  if draw==True:
    canvas.draw_circle01(radius01=.02, center01=point, color='gold', thickness=canvas.SOLID)
  return point  


canvas.clear()    # clear the canvas from the previous drawing

canvas.draw_circle01(1, center01=(0,0), color='scarlet sage', thickness=canvas.SOLID)
for p in range(n):
  point_on_circle(canvas, p, n, draw=True)

canvas.show()  
```

```
def draw_petals(canvas, petals, n):
  m = petals + 1
  canvas.clear()
  canvas.draw_circle01(1, center01=(0,0), color='scarlet sage', thickness=-1)
  for p in range(n):
    point_from = point_on_circle(canvas, p, n , draw=(n<=100))
    point_to   = point_on_circle(canvas, p*m , n)
    canvas.draw_line01(point_from, point_to, color='vanilla custard')
  

draw_petals(canvas, 1, 100)

canvas.show() 
```

```
canvas.download()
```

```
petals_from=0; petals_to=6; frames=100
n = 250

canvas = Canvas_cv(width=400,height=400)
vid = Video_cv(fps=6, frame=canvas)

import numpy as np
for petals in np.linspace(petals_from, petals_to, frames):
  draw_petals(canvas, petals, n)
  vid.write_frame()


vid.end_video()   

# vid.end_video(download=True, show=False) # if big video
```
